```python
climate_change_df = pd.read_csv("../climate_change.csv", parse_dates=['date'], index_col="date")
climate_change_df.info()
```

```python
fig, ax = plt.subplots()
ax.plot(climate_change_df.index, climate_change_df['co2'])
ax.set_xlabel("Date")
ax.set_ylabel("CO2 (ppm)")
plt.show()
```

```python
decade = climate_change_df["1960-01-01":"1970-01-01"]
decade
```

```python
fig, ax = plt.subplots()
ax.plot(climate_change_df.index, climate_change_df['co2'])
ax.plot(climate_change_df.index, climate_change_df['relative_temp'])
ax.set_xlabel("Date")
ax.set_ylabel("CO2 (ppm)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(climate_change_df.index, climate_change_df['co2'])
ax.set_xlabel("Date")
ax.set_ylabel("CO2 (ppm)")
ax2 = ax.twinx()
ax2.plot(climate_change_df.index, climate_change_df['relative_temp'], color="r")
ax2.set_ylabel("Relative Temp")
plt.show()
```

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x,y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.tick_params(colors=color, axis='y')

fig, ax = plt.subplots()
plot_timeseries(ax, climate_change_df.index, climate_change_df['co2'], 'y', 'Date', 'CO2 (ppm)')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(climate_change_df.index, climate_change_df['relative_temp'])
ax.annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2), 
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.set_xlabel("Date")
ax.set_ylabel("CO2 (ppm)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(climate_change_df.index, climate_change_df['co2'])
ax.annotate(">340 ppm", xy=(pd.Timestamp('1980-10-06'), 340), xytext=(pd.Timestamp('2000-10-06'), 320), 
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.set_xlabel("Date")
ax.set_ylabel("CO2 (ppm)")
plt.show()
```

```python
medals_df = pd.read_csv("../medals_by_country_2016.csv", index_col=0)
```

```python
medals_df.info()
```

```python
fig, ax = plt.subplots()
ax.bar(medals_df.index, medals_df['Gold'])
ax.set_xticklabels(medals_df.index, rotation=90)
ax.set_ylabel("Number of Gold medal")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals_df.index, medals_df['Gold'], label='Gold')
ax.bar(medals_df.index, medals_df['Silver'], bottom=medals_df['Gold'], label='Silver')
ax.bar(medals_df.index, medals_df['Bronze'], bottom=medals_df['Gold']+medals_df['Silver'], label='Bronze')
ax.set_xticklabels(medals_df.index, rotation=90)
ax.set_ylabel("Number of Gold medal")
ax.legend()
plt.show()
```

```python
medals_df.head(10)
```

```python
fig, ax = plt.subplots()
ax.hist(medals_df['Gold'], label='Gold', bins=5)
# ax.hist(medals_df['Silver'], label='Silver', bins=10)
ax.set_xlabel("Number of medals won")
ax.set_ylabel("Number of countries")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals_df.index, medals_df['Gold'].mean(), yerr=medals_df['Gold'].std())
ax.set_xticklabels(medals_df.index, rotation=90)
ax.set_ylabel("Number of Gold medal")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.errorbar(climate_change_df.index, climate_change_df['co2'], yerr=climate_change_df['co2'].std())
ax.set_xlabel("Date")
ax.set_ylabel("CO2 (ppm)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot([medals_df['Gold'], medals_df['Silver'], medals_df['Bronze']], label=['Gold', 'Silver', 'Bronze'])
ax.set_xlabel([col for col in medals_df.columns])
ax.legend()
plt.show()
```

```python
decade = climate_change_df["1960-01-01":"1970-01-01"]

fig, ax = plt.subplots()
ax.scatter(climate_change_df['co2'], climate_change_df['relative_temp'], label='Total')
ax.scatter(decade['co2'], decade['relative_temp'], label='1960s')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(climate_change_df['co2'], climate_change_df['relative_temp'],c=climate_change_df.index, label='Total')
ax.legend()
plt.show()
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots()
ax.scatter(climate_change_df['co2'], climate_change_df['relative_temp'],c=climate_change_df.index, label='Total')
ax.legend()
plt.show()
fig.set_size_inches(6,8)
fig.savefig('image.svg', dpi=300)
```
